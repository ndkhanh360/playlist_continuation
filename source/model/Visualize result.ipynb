```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('result.csv')
```

```python
data = pd.melt(df, ['Fraction'])
```

```python
data.columns = ['Fraction', 'Model', 'R-precision']
```

```python
sns.set_style("darkgrid")
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='Fraction', y='R-precision', hue='Model', 
             data=data)
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
plt.title('Performace of different models with different fraction values')
plt.savefig('../../images/results.png', bbox_inches = 'tight')
```
